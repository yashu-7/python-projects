```python
import os
import cv2
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_path = 'dataset'
```

```python
images = f'{dataset_path}\\Images'
captions = f'{dataset_path}\\captions.csv'
print(f"{images}\n{captions}")
```

```python
df = pd.read_csv(captions)
df
```

```python
ref_img = "image-captioning\\dataset\\Images\\1000268201_693b08cb0e.jpg"
```

```python
first_img = df['image'].iloc[0]
first_img_captions = df[df['image'] == first_img]['caption'].tolist()
print(first_img_captions)
print(len(first_img_captions))
```

```python
# plt.figure(figsize=(15,15))
img_path = f'dataset\\Images\\{first_img}'
img = cv2.imread(img_path)
plt.figure(figsize=(25, 20))
for i in range(len(first_img_captions)):
    plt.subplot(1, len(first_img_captions), i + 1)
    plt.imshow(img[:, :, ::-1])
    plt.title(first_img_captions[i])

plt.tight_layout()
plt.show()
```
